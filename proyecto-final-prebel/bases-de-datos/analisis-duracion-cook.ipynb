```python
file_path = r"C:\Users\thomm\Documents\GitHub\modelacion-y-simulacion-5\proyecto-final-prebel\bases-de-datos\datos-discretos.xlsx"
df = pd.read_excel(file_path)
```

```python
# Initialize an empty list to keep track of non-repeated 'Orden'
ordenes = []

# Loop through the 'Orden' column and collect non-repeated values
for i in range(len(df['T Intervalo'])):
    if df['Orden'][i] not in ordenes:
        ordenes.append(df['Orden'][i])

# Create a dictionary from the list of non-repeated 'Orden'
ordenes_dict = {orden: 0 for orden in ordenes}
ordenes = ordenes[1:]
```

```python
for i in range(len(df['T Intervalo'])):
    orden = df['Orden'][i]
    intervalo = df['T Intervalo'][i]

    ordenes_dict[orden] += intervalo

del ordenes_dict['0']
```

```python
tiempos_cook = []
for orden in ordenes:
    tiempos_cook.append(ordenes_dict[orden])
```

```python
# Create a DataFrame with two columns
df_aux = pd.DataFrame({'Ordenes': ordenes, 'Tiempos': tiempos_cook})

df_aux.to_excel('tiempos-cook.xlsx', index=True)
```
